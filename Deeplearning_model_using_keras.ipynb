```python
# Import necessary libraries 
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.optimizers import Adam
from keras.layers import Dense
from keras.losses import mean_squared_error
```

```python
# Load the dataset

data = pd.read_csv('concrete_data.csv')
```

```python
# Viewing the data
data.head(10)
```

```python
# Structure of the data
data.shape
```

```python
# Splitting features variables and target variable

X = data.drop(columns='Strength').values
y = data['Strength'].values
```

```python
def building_and_evaluating_model(X, y, test_size=0.3, epochs=50):
    #Splitting the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=None)
    
    #Building the model
    model = Sequential()
    model.add(Dense(10, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(1))
    
    #Compiling the model with adam optimizer and MSE loss function
    model.compile(optimizer=Adam(), loss='mean_squared_error')
    
    #Train the model
    model.fit(X_train, y_train, epochs=epochs, verbose=0)
    
    #Evaluating model on the test data
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    
    return mse
```

```python
# Repeat the process 50 times to gather MSE values
mse_list = [building_and_evaluating_model(X, y) for _ in range(50)]
```

```python
# Calculate the mean and standard deviation of the MSE values
mse_mean = np.mean(mse_list)
mse_std = np.std(mse_list)
```

```python
print(f'Mean MSE: {mse_mean}')
print(f'Standard Deviation of MSE: {mse_std}')
```
